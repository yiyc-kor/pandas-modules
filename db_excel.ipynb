```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('./data/import.xlsx',
                   names=['hscode_1', 'hscode_2', 'hscode_3', 'hscode_4', 'item_name', 'descriptions', 'related_law'],
                   dtype={'hscode_1': str, 'hscode_2':str, 'hscode_3':str, 'hscode_4':str}).assign(division = 'import')
```

```python
df['hscode'] = df.hscode_1 + df.hscode_2 + df.hscode_3 + df.hscode_4
cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
df = df[cols]
df.apply(lambda x: x.str.strip() if x.dtype == "object" else x).head()
df = df.replace(r'^\s*$', np.nan, regex=True)
df = df.replace(r'\n', ' ', regex=True)
```

```python
df.head()
```

```python
df = df.fillna(method='ffill')
df.head()
```

```python
df.info()
```
