```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data/big_raw_data_6.txt', sep='\t', header=None, 
                 names=["name", "code"], dtype={"name": str, "code": str})
df
```

```python
df = df.loc[(df["code"].str.len() > 5)].reset_index(drop=True)
df
```

```python
df.code = df.code.str.slice(stop=6)
df
```

```python
df.name = df.name.str.replace(r"[.]{3}|[^a-zA-Z\- ]+", " ").str.replace(r"[\.\-]|\b([a-z]{1})\b", "", regex=True)\
                        .str.replace(r"[ ]{2,}", " ").str.strip()
df
```

```python
df = df[(df.name != '') & (df.name != ' ')].reset_index(drop=True)
df
```

```python
df = df.groupby(["name", "code"]).agg('size').reset_index(name="counts")
df
```

```python
df.to_csv("./data/big_prep_data_6.txt", sep='\t', encoding='utf-8', header=False, index=False, columns=["name", "code", "counts"])
```
